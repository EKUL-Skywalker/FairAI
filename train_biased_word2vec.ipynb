```python
from utils.dataset import Dataset
from models.word2vec import Word2Vec
import pickle as pkl
```

```python
ds = Dataset("../simplewiki-20171103-pages-articles-multistream.xml.bz2")
indices = pkl.load(open("dataset.pkl", "rb"))
```

```python
dataset = [ds.lines[i] for i in indices]
model = Word2Vec(window_size=8, dim=25)
model.fit(iid=None, dataset=dataset)
model.save("biased_word2vec.bin")
```

```python
model = Word2Vec()
model.load("biased_word2vec.bin")
```
